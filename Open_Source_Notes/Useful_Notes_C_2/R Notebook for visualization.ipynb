```r
path_op= "/home/qx816/notebooks/weibull_mar7/"

om_output=read.csv(gsub(" ","",paste(path_op,"OM_MODEL_OUTPUT_2019-03-12.csv")),stringsAsFactors=FALSE,sep="|")
```

```r
sapply(om_output,typeof)
```

```r
library(ggplot2)
```

```r
f1 = om_output[om_output$CALC_DATE == "2019-03-07",]
```

```r
nrow(f1)
```

```r
cd <- unique(f1$OM_SCORE_REL_BUCKET)
print(cd)
```

```r
g <- ggplot(f1, aes(OM_SCORE_REL_BUCKET,y = factors(CALC_ID))
g + geom_bar(stat = "identity", width = 0.2, fill="brown")+labs(title="Bar Chart CALC_DATE VS OM_SCORE_REL_BUCKET", 
           subtitle="FOR THE CALC_DATE == 2019-03-07 ")+ theme(axis.text.x = element_text(angle=66, vjust=0.6))
```

```r
f2=om_output[om_output$SOURCE=="TELEMATICS" & om_output$OM_ALGO=="MA_Ribbon" & om_output$CODE=="2398",]
```

```r
g <- ggplot(f2, aes(CALC_DATE,OM_SCORE))
g + geom_point(stat = "identity",color="brown")+ theme(axis.text.x = element_text(angle=66, vjust=0.6))+labs(title="Point Chart for CALC_DATE VS OM_SCORE", 
           subtitle="ON BASIS OF SOURCE== TELEMATICS , OM_ALGO== MA_Ribbon , CODE== 2398 ")
```

```r

f3=om_output[om_output$OM_KEY=='2398-All-All-All-TELEMATICS-X-WARRANTY-COUNT_NBR-MA_Ribbon',] 
 
```

```r
g <- ggplot(f3, aes(CALC_DATE,OM_SCORE,group=1))
g + geom_path(stat = "identity",color="brown")+ theme(axis.text.x = element_text(angle=66, vjust=0.6))+labs(title="Point Chart for CALC_DATE VS OM_SCORE", 
           subtitle="ON BASIS OF SOURCE== TELEMATICS , OM_ALGO== MA_Ribbon , CODE== 2398 ")+geom_point()
```
